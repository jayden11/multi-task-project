```python
import numpy as np
import pandas as pd
import csv
import nltk
```

```python
blood = open('../../data/pubmed_scrapes/blood.fcgi')
human = open('../../data/pubmed_scrapes/human.fcgi')
transcription_factors = open('../../data/pubmed_scrapes/transcription-factors.fcgi')
```

```python
raw_blood = blood.read()
raw_human = human.read()
raw_trans = transcription_factors.read()
```

```python
a = [nltk.wordpunct_tokenize(raw) for raw in [raw_blood[:5000000], raw_human[:5000000], raw_trans[:5000000]]]
a = np.concatenate((a[0],a[1],a[2]))
```

```python
a
```

```python
charar = np.chararray((len(a)))
charar[:] = '.'
chunkar = np.chararray((len(a)))
chunkar[:] = 'O'
```

```python
b = np.vstack((a,charar,chunkar))
```

```python
b = np.transpose(b)
b = pd.DataFrame(b)
b
```

```python
b.to_csv('pubmed_scrape.txt', header=False, sep=' ', index=False)
```
